### Overall Descriptive Statistics

```python
import pandas as pd
import re

# Identify all columns for each dimension across apps
usability_cols = [c for c in df.columns if re.search(r'_U\d+$', c)]
affect_cols    = [c for c in df.columns if re.search(r'_A\d+$', c)]
value_cols     = [c for c in df.columns if re.search(r'_V\d+$', c)]

# Compute averages for each dimension (per respondent)
df['Usability_avg'] = df[usability_cols].mean(axis=1)
df['Affect_avg']    = df[affect_cols].mean(axis=1)
df['UserValue_avg'] = df[value_cols].mean(axis=1)

# Compute descriptive statistics
overall_desc = df[['Usability_avg', 'Affect_avg', 'UserValue_avg']].describe()

# Add range and IQR rows
overall_desc.loc['range'] = overall_desc.loc['max'] - overall_desc.loc['min']
overall_desc.loc['IQR']   = overall_desc.loc['75%'] - overall_desc.loc['25%']

# Rename 50% -> median
overall_desc.rename(index={'50%': 'median'}, inplace=True)

# Reorder rows for clarity (includes IQR)
overall_desc = overall_desc.reindex([
    'count', 'mean', 'std', 'min', '25%', 'median', '75%', 'IQR', 'max', 'range'
])

# Rename columns
overall_desc.columns = ['Usability', 'Affect', 'UserValue']

# Save as CSV
overall_desc.round(3).to_csv('overall_descriptive_stats.csv')

print("Saved overall descriptive statistics to 'overall_descriptive_stats.csv'")
```

### Usability Distribution by App

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

apps = ["AquaSense", "FMS", "FloodWarning"]

# Ensure numeric
for u in range(1, 6):
    for app in apps:
        col = f"{app}_U{u}"
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

titles = {
    1: "U1 — The app is easy to use.",
    2: "U2 — The app is simple, intuitive, and not complex.",
    3: "U3 — The app interface is easy to navigate.",
    4: "U4 — The app assists in completing tasks effortlessly.",
    5: "U5 — The app provides all necessary information."
}

def counts_for_U(u_num):
    rating_values = [1, 2, 3, 4, 5]
    counts = {app: [] for app in apps}
    for r in rating_values:
        for app in apps:
            col = f"{app}_U{u_num}"
            counts[app].append((df[col] == r).sum())
    return rating_values, counts

# --- combined plot with legend for each subplot ---
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(8, 12))

width = 0.22
for idx, u in enumerate(range(1, 6)):
    ratings, counts = counts_for_U(u)
    x = np.arange(len(ratings))
    offsets = {apps[0]: -width, apps[1]: 0.0, apps[2]: width}

    ax = axes[idx]
    for app in apps:
        ax.bar(x + offsets[app], counts[app], width, label=app)

    ax.set_title(titles[u], fontsize=10)
    ax.set_xlabel("Rating")
    ax.set_ylabel("Count")
    ax.set_xticks(x)
    ax.set_xticklabels(ratings)

    # legend for each subplot
    ax.legend(
        title="App",
        frameon=False,
        loc="center left",
        bbox_to_anchor=(1.02, 0.5),
        borderaxespad=0.0
    )

# adjust layout to fit legends
fig.tight_layout()
plt.savefig("usability_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Regression: Predict UX for each App across 3 different scenarios

```python
df = pd.read_csv('ux_dataset_final.csv')
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

apps = ["AquaSense", "FMS", "FloodWarning"]
all_results = []  # collect for CSV

def split_train_eval(X, y, model):
    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )
    # Fit
    model.fit(X_train, y_train)

    # ---- Train RMSE ----
    y_pred_train = model.predict(X_train)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))

    # ---- Test metrics ----
    y_pred_test = model.predict(X_test)
    mse_test = mean_squared_error(y_test, y_pred_test)
    rmse_test = np.sqrt(mse_test)
    r2_test = r2_score(y_test, y_pred_test)

    # ---- Gaps ----
    rmse_gap = rmse_test - rmse_train  # absolute difference (signed)
    rmse_gap_pct = np.nan if rmse_train == 0 else ((rmse_test - rmse_train) / rmse_train) * 100.0

    return rmse_train, mse_test, rmse_test, r2_test, rmse_gap, rmse_gap_pct

for app in apps:
    U_cols = [c for c in df.columns if c.startswith(f"{app}_U")]
    A_cols = [c for c in df.columns if c.startswith(f"{app}_A")]
    V_cols = [c for c in df.columns if c.startswith(f"{app}_V")]
    y_col  = f"{app}_ux"

    scenarios = {
        "Scenario 1 (U only)": U_cols,
        "Scenario 2 (A only)": A_cols,
        "Scenario 3 (V only)": V_cols,
    }

    models = {
        "Linear Regression": LinearRegression(),
        "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
        "SVM (SVR)": make_pipeline(StandardScaler(), SVR(kernel="rbf", C=10.0, epsilon=0.1)),
    }

    for scen_name, feat_cols in scenarios.items():
        X = df[feat_cols].values
        y = df[y_col].values

        rows = []
        for model_name, model in models.items():
            rmse_train, mse_test, rmse_test, r2_test, rmse_gap, rmse_gap_pct = split_train_eval(X, y, model)
            rows.append([model_name, rmse_train, rmse_test, mse_test, r2_test, rmse_gap, rmse_gap_pct])

            all_results.append({
                "App": app,
                "Scenario": scen_name,
                "Model": model_name,
                "Train_RMSE": rmse_train,
                "Test_MSE": mse_test,
                "Test_RMSE": rmse_test,
                "Test_R2": r2_test,
                "RMSE_Gap": rmse_gap,
                "RMSE_Gap_%": rmse_gap_pct
            })

        # Pretty print per scenario (sorted by Test_RMSE)
        table = pd.DataFrame(
            rows, columns=["Model", "Train_RMSE", "Test_RMSE", "MSE", "R²", "RMSE_Gap", "RMSE_Gap_%"]
        ).sort_values("Test_RMSE")

        print("=" * 80)
        print(f"{app} → {scen_name}")
        def fmt(x): return f"{x:.4f}"
        def fmt_pct(x): return "nan" if pd.isna(x) else f"{x:+.2f}%"
        def fmt_signed(x): return f"{x:+.4f}"

        disp = table.copy()
        disp["Train_RMSE"] = disp["Train_RMSE"].map(fmt)
        disp["Test_RMSE"]  = disp["Test_RMSE"].map(fmt)
        disp["MSE"]        = disp["MSE"].map(fmt)
        disp["R²"]         = disp["R²"].map(fmt)
        disp["RMSE_Gap"]   = disp["RMSE_Gap"].map(fmt_signed)
        disp["RMSE_Gap_%"] = disp["RMSE_Gap_%"].map(fmt_pct)
        print(disp.to_string(index=False))

# Save all results to one CSV
results_df = pd.DataFrame(all_results)
results_df.to_csv("model_results_regression.csv", index=False)
print("\nSaved all results to model_results_regression.csv")
```

```python
# ==== Simplified tuning for FloodWarning (linear family models) ====
# Output style matches your initial regression loop

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score

app = "FloodWarning"

# --- feature groups & target
U_cols = [c for c in df.columns if c.startswith(f"{app}_U")]
A_cols = [c for c in df.columns if c.startswith(f"{app}_A")]
V_cols = [c for c in df.columns if c.startswith(f"{app}_V")]
y_col  = f"{app}_ux"

scenarios_fw = {
    "Scenario 1 (U only)": U_cols,
    "Scenario 2 (A only)": A_cols,
    "Scenario 3 (V only)": V_cols,
}

param_grids = {
    "Linear Regression": {},
    "Ridge": {"model__alpha": [0.01, 0.1, 1, 10, 100]},
    "Lasso": {"model__alpha": [0.01, 0.1, 1, 10]},
    "ElasticNet": {"model__alpha": [0.01, 0.1, 1], "model__l1_ratio": [0.2, 0.5, 0.8]},
}

models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(random_state=42),
    "Lasso": Lasso(random_state=42, max_iter=10000),
    "ElasticNet": ElasticNet(random_state=42, max_iter=10000),
}

results_tuned = []

for scen_name, feat_cols in scenarios_fw.items():
    X = df[feat_cols].values
    y = df[y_col].values

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    rows = []
    for model_name, model in models.items():
        pipe = Pipeline([("scaler", StandardScaler()), ("model", model)])
        gs = GridSearchCV(pipe, param_grids[model_name], cv=5, scoring="neg_root_mean_squared_error")
        gs.fit(X_train, y_train)

        best = gs.best_estimator_

        # Train / test metrics
        y_pred_tr = best.predict(X_train)
        y_pred_te = best.predict(X_test)

        rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_tr))
        mse_test   = mean_squared_error(y_test, y_pred_te)
        rmse_test  = np.sqrt(mse_test)
        r2_test    = r2_score(y_test, y_pred_te)

        rmse_gap     = rmse_test - rmse_train
        rmse_gap_pct = np.nan if rmse_train == 0 else ((rmse_test - rmse_train) / rmse_train) * 100.0

        rows.append([model_name, rmse_train, rmse_test, mse_test, r2_test, rmse_gap, rmse_gap_pct])

        results_tuned.append({
            "App": app,
            "Scenario": scen_name,
            "Model": model_name,
            "Best_Params": gs.best_params_,
            "Train_RMSE": rmse_train,
            "Test_MSE": mse_test,
            "Test_RMSE": rmse_test,
            "Test_R2": r2_test,
            "RMSE_Gap": rmse_gap,
            "RMSE_Gap_%": rmse_gap_pct
        })

    # Pretty print sorted by Test_RMSE
    print("=" * 80)
    print(f"{app} → {scen_name} (Linear Family Tuning)")
    table = pd.DataFrame(
        rows, columns=["Model", "Train_RMSE", "Test_RMSE", "MSE", "R²", "RMSE_Gap", "RMSE_Gap_%"]
    ).sort_values("Test_RMSE")

    def fmt(x): return f"{x:.4f}"
    def fmt_pct(x): return "nan" if pd.isna(x) else f"{x:+.2f}%"
    def fmt_signed(x): return f"{x:+.4f}"

    disp = table.copy()
    disp["Train_RMSE"] = disp["Train_RMSE"].map(fmt)
    disp["Test_RMSE"]  = disp["Test_RMSE"].map(fmt)
    disp["MSE"]        = disp["MSE"].map(fmt)
    disp["R²"]         = disp["R²"].map(fmt)
    disp["RMSE_Gap"]   = disp["RMSE_Gap"].map(fmt_signed)
    disp["RMSE_Gap_%"] = disp["RMSE_Gap_%"].map(fmt_pct)
    print(disp.to_string(index=False))

# Save CSV
tuned_df = pd.DataFrame(results_tuned)
tuned_df.to_csv("tuning_FloodWarning_linear_family.csv", index=False)
print("\nSaved tuned results to tuning_FloodWarning_linear_family.csv")
```

```python
import pandas as pd
import numpy as np
from collections import Counter

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# Scenarios (feature subsets)
# -------------------------
scenarios = {
    "U_only": "_U",
    "A_only": "_A",
    "V_only": "_V",
}

all_results = []

for scenario_name, suffix in scenarios.items():
    print("\n" + "#"*90)
    print(f"### Scenario: {scenario_name} (using features containing '{suffix}')")
    print("#"*90)

    # -------------------------
    # 1) Use ONLY scenario features
    # -------------------------
    feat_cols = [c for c in df.columns if suffix in c]   # e.g., *_U*, *_A*, *_V*
    X = df[feat_cols].copy()
    y = df["best_app"].copy()

    # -------------------------
    # 2) Train/Test split
    # -------------------------
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )

    # Adjust SMOTE k for smallest class in train
    min_class_count = min(Counter(y_train).values())
    smote_k = max(1, min(5, min_class_count - 1))
    smote = SMOTE(random_state=42, k_neighbors=smote_k)

    class_order = sorted(y.unique())  # consistent label order

    # -------------------------
    # 3) Models (with SMOTE in-pipeline)
    # -------------------------
    models = {
        "Logistic Regression": ImbPipeline(steps=[
            ('scaler', StandardScaler()),
            ('smote', smote),
            ('clf', LogisticRegression(max_iter=1000, multi_class="ovr", random_state=42))
        ]),
        "Random Forest": ImbPipeline(steps=[
            ('smote', smote),
            ('clf', RandomForestClassifier(n_estimators=200, random_state=42))
        ]),
        "SVM": ImbPipeline(steps=[
            ('scaler', StandardScaler()),
            ('smote', smote),
            ('clf', SVC(probability=True, random_state=42))
        ])
    }

    # Helper: get final estimator's classes_
    def get_model_classes_(fitted_model):
        if hasattr(fitted_model, "named_steps"):
            last_key = list(fitted_model.named_steps.keys())[-1]
            return fitted_model.named_steps[last_key].classes_
        return fitted_model.classes_

    # -------------------------
    # 4) Train, Evaluate, Heatmap CMs
    # -------------------------
    for model_name, model in models.items():
        model.fit(X_train, y_train)

        y_pred_train = model.predict(X_train)
        y_pred_test  = model.predict(X_test)
        y_proba_test = model.predict_proba(X_test)

        acc_train = accuracy_score(y_train, y_pred_train)
        acc_test  = accuracy_score(y_test, y_pred_test)
        acc_gap   = acc_test - acc_train
        precision = precision_score(y_test, y_pred_test, average="weighted", zero_division=0)
        recall    = recall_score(y_test, y_pred_test, average="weighted", zero_division=0)
        f1        = f1_score(y_test, y_pred_test, average="weighted", zero_division=0)

        # Ensure probability columns are ordered to class_order for multiclass AUC
        classes_ = get_model_classes_(model)
        idx = [list(classes_).index(c) for c in class_order]
        y_proba_ordered = y_proba_test[:, idx]
        roc_auc   = roc_auc_score(y_test, y_proba_ordered, labels=class_order, multi_class="ovr", average="macro")

        # ---- print ----
        print("=" * 80)
        print(f"{scenario_name} → {model_name}")
        print(f"Train Acc : {acc_train:.4f}")
        print(f"Test  Acc : {acc_test:.4f}")
        print(f"Acc Gap   : {acc_gap:+.4f}")
        print(f"Precision : {precision:.4f}")
        print(f"Recall    : {recall:.4f}")
        print(f"F1-score  : {f1:.4f}")
        print(f"ROC-AUC   : {roc_auc:.4f}")

        # ---- seaborn CM ----
        cm = confusion_matrix(y_test, y_pred_test, labels=class_order)
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                    xticklabels=class_order, yticklabels=class_order)
        plt.title(f"{scenario_name} → {model_name} Confusion Matrix")
        plt.ylabel("True Label")
        plt.xlabel("Predicted Label")
        plt.tight_layout()
        plt.show()

        # ---- store row ----
        all_results.append({
            "Scenario": scenario_name,
            "Model": model_name,
            "Train Accuracy": acc_train,
            "Test Accuracy": acc_test,
            "Accuracy Gap": acc_gap,
            "Precision (weighted)": precision,
            "Recall (weighted)": recall,
            "F1-score (weighted)": f1,
            "ROC-AUC (macro OVR)": roc_auc
        })

# -------------------------
# 5) Save results
# -------------------------
results_df = pd.DataFrame(all_results)
results_df.to_csv("model_results_classification_scenario.csv", index=False)
print("\n✅ Results saved to 'model_results_classification_scenario.csv'")
```

### Resampling and Predict Best App

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Step 1: Select only UX item-level features (those ending in _U1, _A1, ..., _V5)
ux_features = [col for col in df.columns if any(dim in col for dim in ['_U', '_A', '_V']) and col not in ['AquaSense_ux', 'FMS_ux', 'FloodWarning_ux']]

# Step 2: Define feature matrix X and target y
X = df[ux_features]
y = df['best_app']

# Step 3: Split into train-test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Step 4: Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Step 5 (optional): Show new class distribution
from collections import Counter
print("Original class distribution:", Counter(y_train))
print("After SMOTE:", Counter(y_resampled))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Step 6: Train a Random Forest classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_resampled, y_resampled)

# Step 7: Predict on test set
y_pred = rf.predict(X_test)

# Step 8: Evaluation
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print("Random Forest Evaluation Results")
print(f"Accuracy:  {acc:.4f}")
print(f"F1 Score:  {f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Step 9: Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=rf.classes_)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=rf.classes_, yticklabels=rf.classes_, cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Step 1: Create a pipeline with scaling (Logistic Regression needs features on similar scale)
logreg = make_pipeline(StandardScaler(), LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42))

# Step 2: Train the model
logreg.fit(X_resampled, y_resampled)

# Step 3: Predict on test set
y_pred_logreg = logreg.predict(X_test)

# Step 4: Evaluate the model
acc = accuracy_score(y_test, y_pred_logreg)
f1 = f1_score(y_test, y_pred_logreg, average='weighted')

print("Logistic Regression Evaluation Results")
print(f"Accuracy:  {acc:.4f}")
print(f"F1 Score:  {f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_logreg))

# Step 5: Confusion Matrix
cm = confusion_matrix(y_test, y_pred_logreg, labels=logreg.named_steps['logisticregression'].classes_)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=logreg.named_steps['logisticregression'].classes_, yticklabels=logreg.named_steps['logisticregression'].classes_, cmap='Greens')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC

# Step 1: Create a pipeline with scaling (SVM is sensitive to feature scales)
svm = make_pipeline(StandardScaler(), SVC(kernel='rbf', probability=True, random_state=42))

# Step 2: Train the SVM model
svm.fit(X_resampled, y_resampled)

# Step 3: Predict on the test set
y_pred_svm = svm.predict(X_test)

# Step 4: Evaluate the model
acc = accuracy_score(y_test, y_pred_svm)
f1 = f1_score(y_test, y_pred_svm, average='weighted')

print("SVM Evaluation Results")
print(f"Accuracy:  {acc:.4f}")
print(f"F1 Score:  {f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))

# Step 5: Confusion Matrix
cm = confusion_matrix(y_test, y_pred_svm, labels=svm.named_steps['svc'].classes_)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=svm.named_steps['svc'].classes_, yticklabels=svm.named_steps['svc'].classes_, cmap='Purples')
plt.title('Confusion Matrix - SVM')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

## Using Clustering as target

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. Select AquaSense's U, A, V columns (not including _ux)
aqua_cols = [col for col in df.columns if col.startswith('AquaSense_') and any(x in col for x in ['U', 'A', 'V']) and col != 'AquaSense_ux']
X = df[aqua_cols]

# 2. Standardize (KMeans is sensitive to scale)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply KMeans clustering into 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
df['AquaSense_cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Compute mean score per cluster
cluster_means = df.groupby('AquaSense_cluster')[aqua_cols].mean().mean(axis=1)
print(cluster_means)

# Sort clusters and assign UX ratings
sorted_clusters = cluster_means.sort_values().index
cluster_to_rating = {cluster: rating for rating, cluster in enumerate(sorted_clusters, start=1)}

# Apply mapping
df['AquaSense_ux_clustered'] = df['AquaSense_cluster'].map(cluster_to_rating)
```
